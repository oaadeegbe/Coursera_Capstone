# Segmenting and Clustering Neighborhoods in Toronto

This notebook will be using some known libraries to scrape some specific location (Toronto, Canada) data from wikipedia, wrangle the data, and then parse the scraped data into a dataframe.

### Part One: Scraping, Wrangling and DataFrame

```python
#Import the required libraries

import requests
import pandas as pd
from bs4 import BeautifulSoup
import json
```

```python
#Fetch the html data from provided url
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
results = requests.get(url)


#extract essential table data from the fetched data
soup = BeautifulSoup(results.content)
new = soup.find('table', attrs = {'class': 'wikitable sortable'})
```

```python
#Convert the extracted essential data into a list to be wrangled.

result = []
for each in new.stripped_strings:
    result.append(each)
    
res = result[3:]

for i in range(len(res)):
    if res[i] == 'Not assigned':
        res[i-1] = 'Not assigned'
    else:
        continue


final_res = []
for each in res:
    if each == 'Not assigned':
        pass
    else:
        final_res.append(each)
        
print(final_res)
```

```python
#convert created result list into a dictionary (key-value pairs) of the column headers and row entries

postalCode = []
borough = []
neighborhood = []

for i in range(0, len(final_res), 3):
    postalCode.append(final_res[i])
    borough.append(final_res[i + 1])
    neighborhood.append(final_res[i + 2])

    
#print(postalCode)
#print(borough)
#print(neighborhood)

toronto = {'Postal Code': postalCode, 'Borough': borough, 'Neighborhood': neighborhood}
#print(toronto)
```

```python
toronto_df = pd.DataFrame(toronto)
```

### Part Three: Map Plot

```python
#!conda install -c conda-forge folium=0.5.0 --yes
import folium
from sklearn.cluster import KMeans
import numpy as np 
import matplotlib.cm as cm
import matplotlib.colors as colors
```

#### Grouped the Neighborhoods together relative to their Boroughs to see how they can be clustered.

```python
grouped_df = new_df.groupby('Borough').count()
grouped_df
```

```python
# create map using latitude and longitude values
map_toronto = folium.Map(location=[43.6532, -79.3832], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(new_df['Latitude'], new_df['Longitude'], new_df['Borough'], new_df['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([lat, lng], radius=5, popup=label, color='red', fill=True, fill_color='white', fill_opacity=0.7, parse_html=False).add_to(map_toronto)
    
map_toronto
```

#### Then we carry out kmeans clustering into  10 clusters based on the grouped bo

```python
dummy_df = pd.get_dummies(new_df['Postal Code'])
dummy_df = new_df[['Borough']].join(dummy_df)
dummy_df = dummy_df.groupby('Borough').mean().reset_index()
dummy_df
```

```python
kmeans_group = dummy_df.drop('Borough', axis = 1)

kclusters = 10
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(kmeans_group)
kmeans.labels_
```

```python
dummy_df.insert(0, 'Cluster Labels', kmeans.labels_)
dummy_df
```

```python
ddf = dummy_df[['Cluster Labels', 'Borough']]
ddf
```

```python
fdf = new_df.merge(ddf, how = 'right')
fdf
```

```python
# create map
toronto_clusters = folium.Map(location=[43.6532, -79.3832], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []

for lat, lon, point, cluster in zip(fdf['Latitude'], fdf['Longitude'], fdf['Neighborhood'], fdf['Cluster Labels']):
    label = folium.Popup(str(point) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker([lat, lon], radius=5, popup=label, color=rainbow[cluster-1], fill=True, fill_color=rainbow[cluster-1], fill_opacity=0.9).add_to(toronto_clusters)


toronto_clusters
```
